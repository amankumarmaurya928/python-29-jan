```python
# Q3
'''A DML (data manipulation language) refers to a computer programming language that allows you to add (insert),
   delete (delete), and alter (update) data in a database.
   A DML is typically a sublanguage of a larger database language like SQL,
   with the DML containing some of the language's operators.'''
# INSERT
'''INSERT INTO table_name (column1, column2, column3, ...)
   insert is use to add data in the table the values add as per the order of the column in the table.'''
# example
# if table has colume (c1 int  ,c2 varcur(10),c3floot,c4 boolen)  so add the values as( 42, 'ak',2.3, True) 
# in this 42 added to c1 , 'ak' added to c2 and so-on
mydb = mysql.connector.connect(
  host="localhost",
  user="abc",
  password="password"
)
print(mydb)
mycursor = mydb.cursor()
mycursor.execute("CREATE DATABASE text")
mycursor.execute("CREATE TABLE text.table_name(c1 INT, c2 VARCHAR(50), c3 FLOAT, c4 VARCHAR(40))")
mycursor.execute("INSERT INTO TEST1.test_table value(56,'akm',85.20,'mb')")
mydb.close()


# UPDATE
'''Update command is a data manipulation command which is used to edit the records of a table.
   It may be used to update a single row based on a condition,
   all rows or set of rows based on the condition given by the user.'''
# EXAMPLE
mydb = mysql.connector.connect(
  host="localhost",
  user="abc",
  password="password"
)
print(mydb)
mycursor = mydb.cursor()
mycursor.execute("CREATE DATABASE text")
mycursor.execute("CREATE TABLE text.table_name(c1 INT, c2 VARCHAR(50), c3 FLOAT, c4 VARCHAR(40))")
mycursor.execute("UPDATE TABLE table_name set C1='sneha' where C4='CSE'")
mydb.close()



# DELETE
'''Delete command is a data manipulation command which is used to remove records from a table.
   All records may be removed in one go, or a set of records may be deleted based on a condition.'''
# EXAMPLE
mydb = mysql.connector.connect(
  host="localhost",
  user="abc",
  password="password"
)
print(mydb)
mycursor = mydb.cursor()
mycursor.execute("CREATE DATABASE text")
mycursor.execute("CREATE TABLE text.table_name(c1 INT, c2 VARCHAR(50), c3 FLOAT, c4 VARCHAR(40))")
mycursor.execute("DELETE from table_name where name='sneha'")
mydb.close()
```

```python
# Q4
'''The full form of DQL is Data Query Language.
   DQL is a part of the grouping involved in SQL sub-languages.
   it is a component of SQL statement that allows getting data from the database and imposing order upon it.
It includes the SELECT statement. This command allows getting the data out of the database to perform operations with it.
'''
# SELECT
'''An SQL SELECT statement retrieves records from a database table according to clauses (for example, FROM and WHERE )
   that specify criteria.
   The syntax is: SELECT column1, column2 FROM table1, table2 WHERE column2='value''''
# EXAMPLE
mydb = mysql.connector.connect(
  host="localhost",
  user="abc",
  password="password"
)
print(mydb)
mycursor = mydb.cursor()
mycursor.execute("CREATE DATABASE text")
mycursor.execute("CREATE TABLE text.table_name(c1 INT, c2 VARCHAR(50), c3 FLOAT, c4 VARCHAR(40))")
mycursor.execute("CREATE TABLE text.table_name1(c1 INT, c2 VARCHAR(50), c3 FLOAT, c4 VARCHAR(40))")
mycursor.execute("SELECT c1, c2 FROM table_name, table_name1 WHERE c2='value'")
mydb.close()
```
